How to perform  Merge Operation


```python
import pandas as pd
player = ['player1','player2','player3']
points = [8,9,6]
title = ['Game1','Game2','Game3']
df1 = pd.DataFrame({'Player':player,'Points':points,'Title':title})
df1
```

```python
player = ['player1','player5','player6']
power = ['punch','Kick','Elbow']
title = ['Game1','Game5','Game6']
df2 = pd.DataFrame({'Player':player,'Power':power,'Title':title})
df2
```

```python
# Inner Merge
#df1.merge(df2)...........only perform inner merge
df1.merge(df2,on='Player',how='inner')
```

```python
#left merge
df1.merge(df2,on='Player',how='left')
```

```python
#right merge
df1.merge(df2,on='Player',how='right')
```

```python
#Outer merge
df1.merge(df2,on='Player',how='outer')
```

# how to perform join operation

```python
player = ['Player1','Player2','Player3']
point  = [8,9,6]
title = ['Game1','Game2','Game3']
df3 = pd.DataFrame({'Player':player,'Points':point,'Title':title},index =['L1','L2','L3'])
df3
```

```python
player = ['player1','player5','player6']
power = ['Punch','Kick','Elbow']
title = ['Game1','Game5','Game6']
df4 = pd.DataFrame({'Players':player,'Power':power,'Titles':title},index =['L2','L3','L4'])
df4
```

```python
# column name never be same in join.......@@@@@@@@@@@.........All column name should be different..$$$$$$$$$$$
#inner join
df3.join(df4,how='inner')
```

```python
#left join...........................................
df3.join(df4,how='left')
```

```python
#right join...........................................
df3.join(df4,how='right')
```

```python
#outer join...........................................
df3.join(df4,how='outer')
```

# How to Concatinate  two DataFrame

```python
pd.concat([df3,df4])
```
